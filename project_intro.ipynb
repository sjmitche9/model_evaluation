You may want to search the output for a certain city and get the network id(s). You can then access more info (e.g. stations) about a certain network id using the other endpoint.

```python
# get data for certain network_id
network_id = 'mobibikes' # vancouver
network_json = requests.get("http://api.citybik.es/v2/networks/"+network_id).json()
pprint(network_json)
```

```python
network_json["network"]["stations"][0]
```

```python
import json

with open('van.json', 'w') as f:
  json.dump(network_json, f)
```

Remember that data pulled from the API may be updated at any point (e.g. number of free bikes).

```python
network_json['network']['stations'][0]
```

```python
# creating lists of values from the json
names = [station['name'] for station in network_json['network']['stations']]
longitude = [station['longitude'] for station in network_json['network']['stations']]
latitude = [station['latitude'] for station in network_json['network']['stations']]
# remember to get all of the required columns!
```

```python
import pandas as pd
df_van = pd.DataFrame(
				{
			'name' : names,
            'longitude' : longitude,
            'latitude' : latitude
			}
		)
```

```python
df_van.to_csv('van.csv')
```
